### nonlocal
- local 영역에서 enclosing 영역과 일치하도록 명시

```python
a='global'
def y():
    a='enclosing'
    def z():
        global a
        a=a+'!'
        return a
    return z()
y()  
```

```python
a='global'
def y():
    a='enclosing'
    def z():
        nonlocal a
        a=a+'!'
        return a
    return z()
y()  
```

```python
def y():
    def z():
        # enclosing 영역에 없으면 binding error 
        nonlocal a
        a=a+'!'
        return a
    return z()
y()  
```

```python
def y():
    def z():
        #global a  # 문제 없음
        b=a+'!'
        return b
    return z()
y()  
```
