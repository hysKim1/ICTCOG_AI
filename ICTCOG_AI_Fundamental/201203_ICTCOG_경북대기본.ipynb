```python
plt.subplot(2,2,1)  #state machine 방식
plt.plot([1,2,3])
plt.subplot(2,2,2)
plt.plot([3,2,1])
plt.subplot(224)
plt.subplot(223)
```

```python
fig, ax=plt.subplots(2,2,sharex=True,sharey=False)
```

```python
#혼용 사용
ax1=plt.subplot(2,2,1)  #state machine 방식
plt.plot([1,2,3],'ro')
ax2=plt.subplot(2,2,2, sharex=ax1,sharey=ax1)
plt.plot([3,2,1])
ax3=plt.subplot(224)
ax4=plt.subplot(223)
plt.tight_layout() #자동으로 맞춰줌
```

```python
fig,ax=plt.subplots(2,2,sharex=True, constrained_layout=True) #제약사항 (자동)
```

복잡한 레이아웃 구성
