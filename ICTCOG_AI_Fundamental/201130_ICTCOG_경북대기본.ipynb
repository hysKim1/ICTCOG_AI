```python
import numpy as np
a=np.arange(12).reshape(3,4)
a
```

```python
a[a>4] #boolean indexing
```

```python
a[(0,2),(2,1)] #인덱싱
```

```python
np.all([1,2,1]) #모든 원소가 True(존재론적)
```

```python
np.any([0,0,0]) #원소 하나라도 False(존재록적)
```

```python
np.nonzero([2,0,1]) #0아닌 값 인덱스 반환
```

```python
np.where(a>4) #SQL Where 연관지으면 쉬움
```

```python
np.where((a>3)&(a<5))  # 3<a<5 -> a=4 
```

```python
a[1,0] #상동
```

```python
np.where(True,1,2) #조건이 맞으며 1 아니면 2
```

```python
np.where([True,True], 
         [1,2],[3,4])
```

```python
np.where([True,False], 
         [1,2],[3,4])
```

```python
np.where([[True,False],[True,True]], 
         [[1,2],[3,4]],[[5,6],[7,8]])
```

```python
np.where(a>4,a,0) #4보다 크면 해당 값 작으면 0
```
