- 데이터 표준화 하고나서 knn

```python
data.target_names
```

```python
def std_train(clf,X,y):
    pipe=Pipeline([('std',StandardScaler()),('clf',clf())])
    return pipe.fit(X,y)
```

```python
X=iris.iloc[:,:-1]
y=iris.target
```

```python
std_train(LogisticRegression, X,y)
```

```python
std_train(KNeighborsClassifier,X,y)
```

- 더운 간단하게 사용

```python
def std_train(clf,X,y):
    pipe=Pipeline([('std',StandardScaler()),('clf',clf())])
    return cross_val_score( pipe, X,y,cv=10)
```

```python
std_train(LogisticRegression, X,y)
```

```python
std_train(KNeighborsClassifier,X,y)
```
